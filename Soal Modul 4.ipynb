```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

input data

```python
data = [
    ("Budi", 81.24, 73.66, 95.89, 2023),
    ("Siti", 98.52, 84.86, 88.7, 2024),
    ("Asep", 91.96, 71.03, 79.93, 2022),
    ("Dewi", 87.96, 97.28, 71.91, 2024),
    ("Andi", 74.68, 77.76, 79.33, 2023),
    ("Lestari", 74.68, 89.88, 79.76, 2022),
    ("Joko", 71.74, 79.35, 91.89, 2022),
    ("Rina", 95.99, 85.6, 89.13, 2022),
    ("Fajar", 88.03, 86.4, 96.62, 2024),
    ("Indah", 91.24, 75.55, 84.17, 2023),
    ("Agus", 70.62, 99.09, 73.59, 2022),
    ("Fitri", 99.1, 93.25, 91.4, 2022),
    ("Rudi", 94.97, 98.18, 92.82, 2022),
    ("Maya", 76.37, 96.84, 86.84, 2024),
    ("Hadi", 75.45, 87.94, 93.13, 2024),
    ("Sri", 75.5, 97.66, 84.81, 2023),
    ("Wawan", 79.13, 72.65, 85.68, 2024),
    ("Yuni", 85.74, 75.88, 82.83, 2022),
    ("Bambang", 82.96, 71.36, 70.76, 2023),
    ("Ayu", 78.74, 79.76, 73.24, 2022),
]
```

hitung rata rata

```python
rerata = [] #list kosong untuk menyimpan nilai rata rata
tahun = [] #list kosong untuk menyimpan tahun

for d in data: #loop untuk setiap mahasiswa
    rata = np.mean([d[1], d[2], d[3]]) #menghitung rata rata dari 3 matkul
    rerata.append(rata) #menyimpan hasil rata rata ke list rerata
    tahun.append(d[4]) #menyimpan tahun

```

mengelompokkan

```python
kelompok = {} # menyimpan data nilai berdasarkan angkatan

for t, r in zip(tahun, rerata): # mengelompokkan nilai sesuai tahun
    if t not in kelompok:
        kelompok[t] = []
    kelompok[t].append(r)

nilai_rata = [] # variabel untuk perhitungan rata-rata

for t in sorted(kelompok.keys()):  # mengurutkan tahun

    nilai = kelompok[t] # mengambil nilai per angkatan

    nilai_max = max(nilai)  # nilai terbesar
    nilai_min = min(nilai)  # nilai terkecil
    nilai_rata = sum(nilai)/len(nilai) # menghitung rata-rata

    print(f"Angkatan {t}:")
    print(f"  Maksimum  = {nilai_max:.2f}")
    print(f"  Minimum   = {nilai_min:.2f}")
    print(f"  Rata-rata = {nilai_rata:.2f}")
    print()
```

visualisasi

```python
data_boxplot = [] # menampung data nilai per angkatan untuk boxplot
label = [] # menyimpan label (tahun angkatan)

for t in sorted(kelompok.keys()): # mengurutkan tahun angkatan
    data_boxplot.append(kelompok[t]) # menambahkan data nilai ke list
    label.append(str(t)) # menambahkan label tahun

plt.figure(figsize=(8,5)) # mengatur ukuran grafik

plt.boxplot(data_boxplot, tick_labels=label) # membuat boxplot berdasarkan data dan label

plt.xlabel("Tahun Angkatan")
plt.ylabel("Nilai")
plt.title("Boxplot Nilai per Kelompok Angkatan")

plt.show() # menampilkan grafik
```
