```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph() # membuat graf tidak berarah

edges = [
('Arad','Zerind',75),
('Arad','Timisoara',118),
('Arad','Sibiu',140),
('Zerind','Oradea',71),
('Oradea','Sibiu',151),
('Timisoara','Lugoj',111),
('Lugoj','Mehadia',70),
('Mehadia','Drobeta',75),
('Drobeta','Craiova',120),
('Craiova','Pitesti',138),
('Craiova','Rimnicu Vilcea',146),
('Sibiu','Rimnicu Vilcea',80),
('Sibiu','Fagaras',99),
('Rimnicu Vilcea','Pitesti',97),
('Fagaras','Bucharest',211),
('Pitesti','Bucharest',101),
('Bucharest','Giurgiu',90),
('Bucharest','Urziceni',85),
('Urziceni','Vaslui',142),
('Vaslui','Iasi',92),
('Iasi','Neamt',87),
('Urziceni','Hirsova',98),
('Hirsova','Eforie',86)
]

G.add_weighted_edges_from(edges) # menambahkan edge beserta bobot (jarak)

print("Urutan dari Arad:")

i = 1
for edge in nx.bfs_edges(G, source='Arad'): # traversal BFS dari node Arad
    print(f"{i}. {edge[0]} -> {edge[1]}")
    i += 1
    
pos = {
"Arad":(-6,0),
"Zerind":(-5,2),
"Oradea":(-3,3),
"Sibiu":(-2,0),
"Fagaras":(0,0),
"Rimnicu Vilcea":(-1,-1),
"Pitesti":(0,-2),
"Craiova":(-2,-3),
"Drobeta":(-4,-3),
"Mehadia":(-5,-2),
"Lugoj":(-5,-1),
"Timisoara":(-6,-1),
"Bucharest":(2,-2),
"Giurgiu":(2,-4),
"Urziceni":(4,-2),
"Hirsova":(6,-2),
"Eforie":(7,-3),
"Vaslui":(5,0),
"Iasi":(4,1),
"Neamt":(3,2)
} # posisi node agar tampilan menyerupai peta

labels = nx.get_edge_attributes(G,'weight')   # mengambil bobot tiap edge
plt.figure(figsize=(12,7))                   # mengatur ukuran gambar

nx.draw(G,pos,                               
        with_labels=True,                     
        node_size=2000,                        
        node_color="white",                    
        edgecolors="black",                    
        font_size=9)                           # menggambar graf

nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)  # menampilkan bobot pada edge

plt.title("Graf Peta Kota-kota Besar di Eropa")         # memberi judul
plt.show() # menampilkan graf
```

```python
G = nx.Graph() # membuat graf tidak berarah

edges = [
('Jakarta','Cirebon',327),
('Jakarta','Bandung',270),

('Bandung','Cirebon',120),
('Bandung','Yogyakarta',373),

('Cirebon','Semarang',305),
('Cirebon','Yogyakarta',210),

('Semarang','Yogyakarta',109),
('Semarang','Surakarta',97),
('Semarang','Surabaya',369),

('Yogyakarta','Surakarta',60),
('Surakarta','Malang',370),

('Surabaya','Malang',94)
]

G.add_weighted_edges_from(edges) # menambahkan edge beserta jarak antar kota

print("BFS dari Bandung:")

i = 1
for edge in nx.bfs_edges(G, source='Bandung'): # traversal BFS dimulai dari Bandung
    print(f"{i}. {edge[0]} -> {edge[1]}")
    i += 1
    
pos = {
"Jakarta":(-6,2),
"Bandung":(-5,0),
"Cirebon":(-3,2),
"Semarang":(-1,2),
"Yogyakarta":(-2,0),
"Surakarta":(0,0),
"Surabaya":(2,2),
"Malang":(2,0)
} # posisi node agar menyerupai letak geografis sederhana

labels = nx.get_edge_attributes(G,'weight')   # mengambil bobot tiap edge

plt.figure(figsize=(10,5))                    # mengatur ukuran grafik

nx.draw(G,pos,                               
        with_labels=True,                    
        node_size=2000,                 
        node_color="white",                  
        edgecolors="black",                  
        font_size=10)                        # menggambar graf

nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)  # menampilkan jarak pada edge

plt.title("Graf Kota-kota Besar di Pulau Jawa")         # judul grafik
plt.show() # menampilkan hasil
```
