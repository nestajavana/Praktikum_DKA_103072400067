```python
import networkx as nx
import matplotlib.pyplot as plt

#graf eropa
G1 = nx.Graph() #membuat undirected graf kosong

edges_europe = [ #list edge dan jarak antar kota di eropa
    ('Arad','Zerind',75),
    ('Arad','Timisoara',118),
    ('Arad','Sibiu',140),
    ('Zerind','Oradea',71),
    ('Oradea','Sibiu',151),
    ('Timisoara','Lugoj',111),
    ('Lugoj','Mehadia',70),
    ('Mehadia','Drobeta',75),
    ('Drobeta','Craiova',120),
    ('Craiova','Pitesti',138),
    ('Craiova','Rimnicu Vilcea',146),
    ('Sibiu','Rimnicu Vilcea',80),
    ('Sibiu','Fagaras',99),
    ('Rimnicu Vilcea','Pitesti',97),
    ('Fagaras','Bucharest',211),
    ('Pitesti','Bucharest',101),
    ('Bucharest','Giurgiu',90),
    ('Bucharest','Urziceni',85),
    ('Urziceni','Vaslui',142),
    ('Vaslui','Iasi',92),
    ('Iasi','Neamt',87),
    ('Urziceni','Hirsova',98),
    ('Hirsova','Eforie',86)
]

G1.add_weighted_edges_from(edges_europe) #menambahkan edge berbobot ke graf

# UCS = Dijkstra
path_europe = nx.dijkstra_path(G1, source='Arad', target='Bucharest', weight='weight') #mencari jalur biaya minimum dari Arad ke Bucharest
cost_europe = nx.dijkstra_path_length(G1, source='Arad', target='Bucharest', weight='weight') #menghitung total biaya jalur tersebut

print("UCS Graf Eropa") #menampilkan hasil pencarian
print("Jalur :", " -> ".join(path_europe)) #menampilkan urutan kota yang dilalui
print("Total biaya :", cost_europe) #menampilkan total jarak

#visualisasi graf eropa
pos1 = { #posisi tiap node agar graf mudah dibaca
    "Arad":(-6,0),"Zerind":(-5,2),"Oradea":(-3,3),
    "Sibiu":(-2,0),"Fagaras":(0,0),"Rimnicu Vilcea":(-1,-1),
    "Pitesti":(0,-2),"Craiova":(-2,-3),"Drobeta":(-4,-3),
    "Mehadia":(-5,-2),"Lugoj":(-5,-1),"Timisoara":(-6,-1),
    "Bucharest":(2,-2),"Giurgiu":(2,-4),"Urziceni":(4,-2),
    "Hirsova":(6,-2),"Eforie":(7,-3),"Vaslui":(5,0),
    "Iasi":(4,1),"Neamt":(3,2)
}

labels1 = nx.get_edge_attributes(G1,'weight') #mengambil bobot tiap edge

plt.figure(figsize=(12,7)) #membuat kanvas gambar
nx.draw(G1, pos1, with_labels=True, node_size=2000,
        node_color="white", edgecolors="black")

nx.draw_networkx_edge_labels(G1, pos1, edge_labels=labels1) #menampilkan bobot edge

# highlight jalur UCS
edges_path1 = list(zip(path_europe, path_europe[1:])) #membuat pasangan edge dari jalur hasil UCS
nx.draw_networkx_edges(G1, pos1, edgelist=edges_path1, width=3) #menebalkan jalur hasil pencarian

plt.title("Graf Eropa (UCS)")
plt.show()


#graf jawa
G2 = nx.Graph() #membuat undirected graf kosong

edges_jawa = [ #list edge dan jarak antar kota di jawa
    ('Jakarta','Cirebon',327),
    ('Jakarta','Bandung',270),
    ('Bandung','Cirebon',120),
    ('Bandung','Yogyakarta',373),
    ('Cirebon','Semarang',305),
    ('Cirebon','Yogyakarta',210),
    ('Semarang','Yogyakarta',109),
    ('Semarang','Surakarta',97),
    ('Semarang','Surabaya',369),
    ('Yogyakarta','Surakarta',60),
    ('Surakarta','Malang',370),
    ('Surabaya','Malang',94)
]

G2.add_weighted_edges_from(edges_jawa) #menambahkan edge berbobot ke graf

path_jawa = nx.dijkstra_path(G2, source='Bandung', target='Malang', weight='weight') #mencari jalur biaya minimum dari Bandung ke Malang
cost_jawa = nx.dijkstra_path_length(G2, source='Bandung', target='Malang', weight='weight') #menghitung total biaya jalur tersebut

print("\nUCS Graf Jawa")
print("Jalur :", " -> ".join(path_jawa)) #menampilkan jalur yang dilalui
print("Total biaya :", cost_jawa) #menampilkan total jarak

#visualisasi graf jawa
pos2 = { #posisi tiap node
    "Jakarta":(-6,2),
    "Bandung":(-5,0),
    "Cirebon":(-3,2),
    "Semarang":(-1,2),
    "Yogyakarta":(-2,0),
    "Surakarta":(0,0),
    "Surabaya":(2,2),
    "Malang":(2,0)
}

labels2 = nx.get_edge_attributes(G2,'weight') #mengambil bobot edge

plt.figure(figsize=(10,5)) #membuat kanvas
nx.draw(G2, pos2, with_labels=True, node_size=2000,
        node_color="white", edgecolors="black")

nx.draw_networkx_edge_labels(G2, pos2, edge_labels=labels2) #menampilkan bobot

# highlight jalur UCS
edges_path2 = list(zip(path_jawa, path_jawa[1:])) #membuat edge dari jalur hasil UCS
nx.draw_networkx_edges(G2, pos2, edgelist=edges_path2, width=3) #menebalkan jalur

plt.title("Graf Jawa (UCS)")
plt.show() #menampilkan graf
```
